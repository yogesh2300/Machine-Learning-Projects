**RANDOM FOREST CLASSIFIER**

```python
import pandas as pd
import numpy as np
```

```python
url ="https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
iris = pd.read_csv(url, names=['sepal length','sepal width','petal length','petal width','target'])

print(iris.target)

```

```python
#Dataframe of iris dataset

df=pd.DataFrame(iris)
df.head()
```

```python
iris
```

```python
df['target']=iris.target
print(iris.target)
```

```python
#traning and testing

from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(df.drop(['target'], axis='columns'), iris.target, test_size = 0.2, random_state = 1)

X_train.shape
```

```python
Y_train.shape
```

```python
# Import random forest classifier

from sklearn.ensemble import RandomForestClassifier

model= RandomForestClassifier(n_estimators=40)
model.fit(X_train,Y_train)
model.score(X_test,Y_test)
```

```python
Y_pred =model.predict(X_test)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test,Y_pred)
cm
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.heatmap(cm, annot = True)
plt.xlabel("Predicted")
plt.ylabel("Truth")
plt.show()
```
